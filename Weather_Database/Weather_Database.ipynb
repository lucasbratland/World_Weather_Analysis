```python
import requests
import pandas as pd
from citipy import citipy
import numpy as np
import time

from config import weather_api_key
```

```python
#Create a set of 2000 random latitudes and longitudes
lat = np.random.uniform(-90, 90, size = 2000)
lng = np.random.uniform(-180, 180, size = 2000)
lat_lng = zip(lat, lng)
lat_lng
```

```python
#Create a list of latitudes and longitudes
coords = list(lat_lng)
```

```python
#Create a list to hold the cities
cities = []

#Identify the nearest city for each latitude and longitude
for coord in coords:
    city = citipy.nearest_city(coord[0], coord[1]).city_name
    
    #If city is unique add it to the list
    if city not in cities:
        cities.append(city)
        
#Print the city count
len(cities)
```

```python
#Retrieve data from OPenWeatherMap API
#Create and empty list
city_data = []

#Create base URL
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

#Create Counter to make sure we don't over load our requests
set_count = 1

#Loop through all the cities in our list
for i, city in enumerate(cities):
    
    #Group cities in sets of 50
    if (i % 50 == 0 and i > 50):
        set_count += 1
        time.sleep(60)
        
    #Create each endpoint URL with each city
    city_url = base_url + "&q=" +city.replace(" ","+")
    
    #RUn an API request for each of the cities
    try:
        #Parse the JSON request and retrieve data
        city_weather = requests.get(city_url).json()
        
        #Parse the needed data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_temp = city_weather["main"]["temp_max"]
        city_humid = city_weather["main"]["humidity"]
        city_cloud = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_desc = city_weather["weather"][0]["description"]
        city_country = city_weather["sys"]["country"]
        
        #Append to city_data
        city_data.append({"City": city.title(),
                          "Country" : city_country,
                         "Lat" : city_lat,
                         "Lng" : city_lng,
                         "Max Temp" : city_temp,
                         "Humidity" : city_humid,
                         "Cloudiness" : city_cloud,
                         "Wind Speed" : city_wind,
                         "Current Description" : city_desc})
    except:
        print(f"City not found. Skipping..")
        pass

        
```

```python
#Create a DataFrame
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
#Export to a csv
output_data_file = "WeatherPy_Database.csv"
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
